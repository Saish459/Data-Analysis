```python
#Importing Libraries
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
print(sns.get_dataset_names())
```

```python
#loading dataset
df = sns.load_dataset("diamonds")
df[:5]
```

```python
df.shape
```

```python
df.describe()
```

```python
sample = df.sample(3000)
```

```python
sample.head()
```

```python
sns.histplot(data=sample, x="price")
```

```python
sns.histplot(data=sample, x="carat", stat = "percent")
```

```python
sns.kdeplot(data=sample, x="table")
```

```python
sample.head()
```

```python
sns.lineplot(data=sample[:20], x='x',y='y')
```

```python
sns.scatterplot(data=sample[:], x='carat', y='price', hue='x')
```

```python
sns.pairplot(sample[["price", "carat", "table", "depth"]])
```

```python
df.corr()
```

```python
#Absolute value of correlation 
#[0-0.3] as weak
#[0.3-0.7] as moderate
#[0.7-1.0] as strong correlation
```

```python
corr_matrix = df.corr()
```

```python
corr_matrix.head()
```

```python
sns.heatmap(corr_matrix, annot=True)
```

```python
g = sns.PairGrid(sample[["price", "carat", "depth"]])
g.map(sns.scatterplot)
```

```python
g = sns.PairGrid(sample[["price", "carat", "depth"]])
g.map_lower(sns.scatterplot)
g.map_upper(sns.kdeplot)

g.map_diag(sns.histplot)
```
