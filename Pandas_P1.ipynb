Replacing data in column using ***.mask()***

```python
# Let's replace data in column(embark_town) from Southampton to SA

dataframe['embark_town'].mask(dataframe['embark_town'] ==  'Southampton', 'SA')
dataframe.head()

#In the output, the changes are not updated. So to update the changes we use 'inplace'=True
#Let's see that in next cell
```

```python
dataframe['embark_town'].mask(dataframe['embark_town'] == 'Southampton', 'SA', inplace = True)
dataframe.head()

#Now it's updated !!
```


In 'survived' column:  
*   1 means survived
*   0 means Not survived

Let's change into 
*   1 as YES
*    0 as NO





```python
dataframe['survived'].mask(dataframe['survived'] == 1, 'YES', inplace =True)
dataframe['survived'].mask(dataframe['survived'] == 0, 'NO', inplace =True)
dataframe.head()
```
